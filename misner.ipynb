```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def map_points(t, x):
    phi = np.pi/20
    return t*np.cosh(phi) + x*np.sinh(phi), x*np.cosh(phi) + t*np.sinh(phi) 
```

```python
num_pts = 100
x = np.arange(0, num_pts)

for t_coord in range(0,100):
    # construct constant t-grid
    t = np.full(num_pts, t_coord)
    t_boost, x_boost = map_points(t, x)
    plt.plot(x,t)
    plt.scatter(x_boost, t_boost)
```
